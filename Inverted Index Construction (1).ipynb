#### Inverted Index Construction
First remove stop words, and punctuation, and apply Porter's stemming algorithm to the 3 documents Show the resulting documents.
Create an inverted index of the three documents, including the dictionary and the postings. The dictionary should also contain (for each term) statistics such as total number of occurrences in the collection and the document frequency. 
The postings for each term should contain the document ids and the term frequencies (depict multiple postings for a term as a linked list).

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import PorterStemmer
```

```python
Doc1="Glimpse is an indexing and query system that allows for search through a file system or document collection quickly. Glimpse is the default search engine in a larger information retrieval system. It has also been used as part of some web based search engines."
```

```python
Doc2="The main processes in an retrieval system are document indexing, query processing, query evaluation and relevance feedback. Among these, efficient updating of the index is critical in large scale systems."
```

```python
Doc3="Clusters are created from short snippets of documents retrieved by web search engines which are as good as clusters created from the full text of web documents."
```

```python
stemmer=PorterStemmer()
def doc_prepare(doc):
    doc=doc.lower()
    tokens=[w for w in word_tokenize(doc) if w not in stopwords.words("english")]
    tokens=[stemmer.stem(w) for w in tokens]
    tokens=[w for w in tokens if w.isalpha()]
    return tokens
```

#### a)First remove stop words, and punctuation, and apply Porter's stemming algorithm to the 3 documents,Show the resulting documents.

```python
Doc1= doc_prepare(Doc1)
Doc2= doc_prepare(Doc2)
Doc3=doc_prepare(Doc3)
```

```python
print("Doc1= ",Doc1,"\n\nDoc2= ",(Doc2),"\n\nDoc3= ",(Doc3))
```

#### b) Create an inverted index of the three documents, including the dictionary and the postings. The dictionary should also contain (for each term) statistics such as total number of occurrences in the collection and the document frequency. The postings for each term should contain the document ids and the term frequencies 

```python
doc=Doc1+Doc2+Doc3
```

```python
unique=[]
total_freq=[]
for key in sorted(nltk.FreqDist(doc).items()):
    unique.append(key[0])
    total_freq.append(key[1])
```

```python
doc_freq=[]

for u in unique:
    count=0
    if u in Doc1:
        count=count+1
    if u in Doc2:
        count=count+1
    if u in Doc3:
        count=count+1
    doc_freq.append((count))
       
```

```python
posting=[]
for u in unique:
    count1=0
    count2=0
    count3=0
    if u in Doc1:
        count1=count1+(nltk.FreqDist(Doc1)[u])
    if u in Doc2:
        count2=count2+(nltk.FreqDist(Doc2)[u])
    if u in Doc3:
        count3=count3+(nltk.FreqDist(Doc3)[u])
    
    posting.append(("Doc 1= ",count1,"Doc 2= ",count2,"Doc 3= ",count3))
  
```

```python
import pandas as pd
```

```python
pd.DataFrame(unique,columns=["terms"])
pd.DataFrame(total_freq,columns=["total term freq"])
pd.DataFrame(doc_freq,columns=["doc freq"])
tb=pd.concat([pd.concat([pd.DataFrame(unique,columns=["terms"]),pd.DataFrame(total_freq,columns=["total term freq"])],axis=1),pd.DataFrame(doc_freq,columns=["doc freq"])],axis=1)
```

```python
pd.concat([tb,pd.DataFrame(posting,columns=["Doc#","freq","Doc#","freq","Doc#","freq"])],axis=1)
```

#### Ngrams: Using bigrams, which of the following terms is more likely to be considered equivalent to the term "informational"?  Justify your answer by using Dice's coefficient to measure similarity of each term to "informational".

```python
# return bigram string
def bigram(w):
    bigram=[]
    for i in range(0,len(w)):
        if i == 0:
            bigram.append(w[:i+2])
        elif i == len(w)-1:
            break
        else:
            bigram.append(w[i:i+2])
    return bigram
```

```python
q="informational"
first="informal"
second="formalization"
```

```python
print("informational= ",bigram(q), "\ninformal= ",bigram(first), "\nformalization= ",bigram(second))
```

#### Dice’s coefficient = 2c/(a+b) 
The dice similarity between “informational” and “informal” = 0.74,  (a,b,c) = (12, 7, 7) 
 
The dice similarity between “informational” and “formalization” = 0.75, 
(a,b,c) = (12, 12, 9) 
According to the calculation above, it seems that the word “informational” and “f ormalization” are more similar to each other in meaning compare to “information al” and “informal”. 


```python
# use Dice’s coefficient to compute “similarity” for pairs of words”

def dice_similarity(w1,w2):
    a=len(set(w1))
    b=len(set(w2))
    c=len(w1+w2)-len(set(w1+w2))
    return (a,b,c),(2*c)/(a+b)
```

```python
dice_similarity(bigram(q),bigram(first))
```

```python
dice_similarity(bigram(q),bigram(second))
```
